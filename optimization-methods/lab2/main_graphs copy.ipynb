```python
import matplotlib.pyplot as plt
import pandas as pd

# Чтение данных из файлов
traj_cd_rosenbrock = pd.read_csv("traj_cd_rosenbrock.csv")
traj_gs_rosenbrock = pd.read_csv("traj_gs_rosenbrock.csv")
traj_hj_rosenbrock = pd.read_csv("traj_hj_rosenbrock.csv")

# Данные для других функций (например, Швефель и Растригин)
traj_cd_schwefel = pd.read_csv("traj_cd_schwefel.csv")
traj_gs_schwefel = pd.read_csv("traj_gs_schwefel.csv")
traj_hj_schwefel = pd.read_csv("traj_hj_schwefel.csv")

traj_cd_rastrigin = pd.read_csv("traj_cd_rastrigin.csv")
traj_gs_rastrigin = pd.read_csv("traj_gs_rastrigin.csv")
traj_hj_rastrigin = pd.read_csv("traj_hj_rastrigin.csv")

# Функция для отрисовки графиков
def plot_trajectory(traj_cd, traj_gs, traj_hj, title):
    plt.figure(figsize=(15, 5))
    
    # Покоординатный спуск
    plt.subplot(1, 3, 1)  # 1 строка, 3 колонки, 1-й график
    plt.plot(traj_cd['x1'], traj_cd['x2'], marker='o', linestyle='-', color='b')
    plt.title("Coordinate Descent")
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.grid(True)
    
    # Гаусс-Зейдель
    plt.subplot(1, 3, 2)  # 1 строка, 3 колонки, 2-й график
    plt.plot(traj_gs['x1'], traj_gs['x2'], marker='o', linestyle='-', color='g')
    plt.title("Gauss-Seidel")
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.grid(True)
    
    # Хук-Дживс
    plt.subplot(1, 3, 3)  # 1 строка, 3 колонки, 3-й график
    plt.plot(traj_hj['x1'], traj_hj['x2'], marker='o', linestyle='-', color='r')
    plt.title("Hooke-Jeeves")
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.grid(True)
    
    # Общий заголовок
    plt.suptitle(title)
    
    # Показать график
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)  # Чтобы заголовок не перекрывался
    plt.show()

# Отображаем для каждой функции
plot_trajectory(traj_cd_rosenbrock, traj_gs_rosenbrock, traj_hj_rosenbrock, "Rosenbrock Function")
plot_trajectory(traj_cd_schwefel, traj_gs_schwefel, traj_hj_schwefel, "Schwefel Function")
plot_trajectory(traj_cd_rastrigin, traj_gs_rastrigin, traj_hj_rastrigin, "Rastrigin Function")
```

```julia
using Plots

f(x, y) = x^2 + y^2 

x_vals = -3:0.1:3
y_vals = -3:0.1:3

grad_prev = [2, 2] 
grad_curr = [2, 1]   
grad_diff = grad_curr - grad_prev
beta = dot(grad_curr, grad_diff) / dot(grad_prev, grad_prev) 

search_dir = -(-grad_curr + beta * grad_prev)

contour(x_vals, y_vals, (x, y) -> f(x, y), levels=levels, xlabel="x", ylabel="y", colorbar=false, size=(500, 500))

scatter!([0, grad_prev[1]], [0, grad_prev[2]], markersize=6, markershape=:circle, markercolor=:blue, label="f_{k-1}")
scatter!([0, grad_curr[1]], [0, grad_curr[2]], markersize=6, markershape=:circle, markercolor=:red, label="f_k")
scatter!([grad_curr[1], grad_curr[1] + grad_diff[1]], [grad_curr[2], grad_curr[2] + grad_diff[2]], markersize=6, markershape=:circle, markercolor=:green, label="f_k - f_{k-1}")
scatter!([0, search_dir[1]], [0, search_dir[2]], markersize=6, markershape=:circle, markercolor=:black, label="  ( )")

plot!([0, grad_prev[1]], [0, grad_prev[2]], label="", color=:blue, linewidth=2)
plot!([0, grad_curr[1]], [0, grad_curr[2]], label="", color=:red, linewidth=2)
plot!([grad_curr[1], grad_curr[1] + grad_diff[1]], [grad_curr[2], grad_curr[2] + grad_diff[2]], label="", color=:green, linewidth=2)
plot!([0, search_dir[1]], [0, search_dir[2]], label="", color=:black, linestyle=:dash, linewidth=2)

display(plot!())
```

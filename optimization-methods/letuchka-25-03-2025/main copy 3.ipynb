```julia
using Plots

#       x1 >= 0  x2 >= 0
function region_condition(x1, x2)
    return -x1 + x2 >= -1 && x1 - 2 * x2 <= 1 && x1 >= 0 && x2 >= 0
end

#    x1  x2
x1_vals = LinRange(0, 3, 100)  #  x1 >= 0
x2_vals = LinRange(0, 3, 100)  #  x2 >= 0

#   ,   
valid_points = []
for x1 in x1_vals
    for x2 in x2_vals
        if region_condition(x1, x2)
            push!(valid_points, (x1, x2))
        end
    end
end

#   x1  x2  
valid_x1, valid_x2 = getindex.(valid_points, 1), getindex.(valid_points, 2)

#  
p = plot()
plot!(p, x1_vals, -x1_vals .+ 1, label="-x1 + x2 >= -1", linewidth=2, linestyle=:dash, color=:blue)  #  1
plot!(p, x1_vals, (x1_vals .- 1) ./ 2, label="x1 - 2x2 <= 1", linewidth=2, linestyle=:dot, color=:red)  #  2

#   
scatter!(p, valid_x1, valid_x2, label="Region", color=:green, alpha=0.3)  #  

#  
display(p)
```

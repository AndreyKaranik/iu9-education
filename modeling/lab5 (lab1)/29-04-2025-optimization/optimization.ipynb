```python
import numpy as np

def f(x):
    """Целевая функция f(x1,x2) = x1^2 + x2^2"""
    return x[0]**2 + x[1]**2

class Particle:
    def __init__(self, id, position, velocity):
        self.id = id
        self.position = np.array(position, dtype=float)
        self.velocity = np.array(velocity, dtype=float)
        self.best_position = self.position.copy()
        self.best_value = f(position)
    
    def update(self, phi_p, phi_g, r_p, r_g, global_best):
        """Обновление скорости и положения частицы"""
        print(f"\nОбновление частицы {self.id}:")

        cognitive = phi_p * r_p * (self.best_position - self.position)
        social = phi_g * r_g * (global_best - self.position)
        
        print(f"  Когнитивная составляющая: {phi_p} * {r_p} * ({self.best_position} - {self.position}) = {cognitive}")
        print(f"  Социальная составляющая: {phi_g} * {r_g} * ({global_best} - {self.position}) = {social}")
        
        new_velocity = self.velocity + cognitive + social
        print(f"  Новая скорость: {self.velocity} + {cognitive} + {social} = {new_velocity}")

        new_position = self.position + new_velocity
        print(f"  Новое положение: {self.position} + {new_velocity} = {new_position}")
        
        self.velocity = new_velocity
        self.position = new_position

        current_value = f(new_position)
        if current_value < self.best_value:
            self.best_position = new_position.copy()
            self.best_value = current_value
            print(f"  Новое лучшее положение частицы: {self.best_position}, f={current_value:.4f}")

def pso(particles, phi_p=1.0, phi_g=1.0, r_p=0.5, r_g=0.5, max_iter=4):
    """Алгоритм оптимизации роем частиц"""
    global_best = None
    global_best_value = float('inf')
    for p in particles:
        if p.best_value < global_best_value:
            global_best_value = p.best_value
            global_best = p.best_position.copy()
    
    print("=== Начальные значения ===")
    print(f"Глобальное лучшее положение: {global_best}, f={global_best_value:.4f}")
    for p in particles:
        print(f"Частица {p.id}: положение={p.position}, скорость={p.velocity}, лучшее={p.best_position}, f={p.best_value:.4f}")
    
    for iteration in range(max_iter):
        print(f"\n=== Итерация {iteration + 1} ===")
        print(f"Текущее глобальное лучшее: {global_best}, f={global_best_value:.4f}")
        
        for p in particles:
            p.update(phi_p, phi_g, r_p, r_g, global_best)
        
        for p in particles:
            if p.best_value < global_best_value:
                global_best_value = p.best_value
                global_best = p.best_position.copy()
                print(f"\nОбновлено глобальное лучшее: {global_best}, f={global_best_value:.4f}")
    
    return global_best, global_best_value

particles = [
    Particle(1, [1.0, 0.0], [1.0, 0.0]),
    Particle(2, [0.0, 4.0], [0.0, 1.0]),
    Particle(3, [-2.12, -2.12], [-0.7, -0.7])
]

best_pos, best_val = pso(particles, phi_p=1.5, phi_g=1.5, r_p=0.5, r_g=0.5)

print("\n=== Финальный результат ===")
print(f"Лучшее найденное решение: {best_pos}, f={best_val:.4f}")
```

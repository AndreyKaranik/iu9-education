```python
import numpy as np

def f(x1, x2):
    """Целевая функция f(x1,x2) = x1^2 + x2^2"""
    return x1**2 + x2**2

class Particle:
    def __init__(self, id, x1, x2, v1, v2):
        self.id = id
        self.x1 = x1  # Положение по x1
        self.x2 = x2  # Положение по x2
        self.v1 = v1  # Скорость по x1
        self.v2 = v2  # Скорость по x2
        self.best_x1 = x1  # Лучшее положение по x1
        self.best_x2 = x2  # Лучшее положение по x2
        self.best_value = f(x1, x2)  # Лучшее значение функции
    
    def update(self, phi_p, phi_g, r_p, r_g, global_best_x1, global_best_x2):
        """Обновление скорости и положения частицы"""
        print(f"\n--- Обновление частицы {self.id} ---")
        print(f"Текущее положение: (x1={self.x1:.4f}, x2={self.x2:.4f})")
        print(f"Текущая скорость: (v1={self.v1:.4f}, v2={self.v2:.4f})")
        
        # Вычисляем когнитивную составляющую
        cognitive_v1 = phi_p * r_p * (self.best_x1 - self.x1)
        cognitive_v2 = phi_p * r_p * (self.best_x2 - self.x2)
        print(f"\nКогнитивная составляющая:")
        print(f"v1: {phi_p}*{r_p}*({self.best_x1:.4f}-{self.x1:.4f}) = {cognitive_v1:.4f}")
        print(f"v2: {phi_p}*{r_p}*({self.best_x2:.4f}-{self.x2:.4f}) = {cognitive_v2:.4f}")
        
        # Вычисляем социальную составляющую
        social_v1 = phi_g * r_g * (global_best_x1 - self.x1)
        social_v2 = phi_g * r_g * (global_best_x2 - self.x2)
        print(f"\nСоциальная составляющая:")
        print(f"v1: {phi_g}*{r_g}*({global_best_x1:.4f}-{self.x1:.4f}) = {social_v1:.4f}")
        print(f"v2: {phi_g}*{r_g}*({global_best_x2:.4f}-{self.x2:.4f}) = {social_v2:.4f}")
        
        # Обновляем скорость
        new_v1 = self.v1 + cognitive_v1 + social_v1
        new_v2 = self.v2 + cognitive_v2 + social_v2
        print(f"\nНовая скорость:")
        print(f"v1: {self.v1:.4f} + {cognitive_v1:.4f} + {social_v1:.4f} = {new_v1:.4f}")
        print(f"v2: {self.v2:.4f} + {cognitive_v2:.4f} + {social_v2:.4f} = {new_v2:.4f}")
        
        # Обновляем положение
        new_x1 = self.x1 + new_v1
        new_x2 = self.x2 + new_v2
        print(f"\nНовое положение:")
        print(f"x1: {self.x1:.4f} + {new_v1:.4f} = {new_x1:.4f}")
        print(f"x2: {self.x2:.4f} + {new_v2:.4f} = {new_x2:.4f}")
        
        self.v1, self.v2 = new_v1, new_v2
        self.x1, self.x2 = new_x1, new_x2
        
        # Обновляем лучшее положение частицы
        current_value = f(new_x1, new_x2)
        if current_value < self.best_value:
            self.best_x1, self.best_x2 = new_x1, new_x2
            self.best_value = current_value
            print(f"\nОбновлено лучшее положение частицы: (x1={self.best_x1:.4f}, x2={self.best_x2:.4f}), f={current_value:.4f}")

def pso(particles, phi_p=1.0, phi_g=1.0, r_p=0.5, r_g=0.5, max_iter=4):
    """Алгоритм оптимизации роем частиц"""
    # Инициализация глобального лучшего положения
    global_best_x1, global_best_x2 = None, None
    global_best_value = float('inf')
    
    for p in particles:
        if p.best_value < global_best_value:
            global_best_x1, global_best_x2 = p.best_x1, p.best_x2
            global_best_value = p.best_value
    
    print("=== НАЧАЛЬНЫЕ ЗНАЧЕНИЯ ===")
    print(f"Глобальное лучшее: (x1={global_best_x1:.4f}, x2={global_best_x2:.4f}), f={global_best_value:.4f}")
    for p in particles:
        print(f"\nЧастица {p.id}:")
        print(f"Положение: (x1={p.x1:.4f}, x2={p.x2:.4f})")
        print(f"Скорость: (v1={p.v1:.4f}, v2={p.v2:.4f})")
        print(f"Лучшее положение: (x1={p.best_x1:.4f}, x2={p.best_x2:.4f}), f={p.best_value:.4f}")
    
    for iteration in range(max_iter):
        print(f"\n\n=== ИТЕРАЦИЯ {iteration + 1} ===")
        print(f"Текущее глобальное лучшее: (x1={global_best_x1:.4f}, x2={global_best_x2:.4f}), f={global_best_value:.4f}")
        
        # Обновляем все частицы
        for p in particles:
            p.update(phi_p, phi_g, r_p, r_g, global_best_x1, global_best_x2)
        
        # Обновляем глобальное лучшее положение
        for p in particles:
            if p.best_value < global_best_value:
                global_best_x1, global_best_x2 = p.best_x1, p.best_x2
                global_best_value = p.best_value
                print(f"\n!!! ОБНОВЛЕНО ГЛОБАЛЬНОЕ ЛУЧШЕЕ !!!")
                print(f"Новое глобальное лучшее: (x1={global_best_x1:.4f}, x2={global_best_x2:.4f}), f={global_best_value:.4f}")
    
    return (global_best_x1, global_best_x2), global_best_value

# Инициализация частиц
particles = [
    Particle(1, x1=1.0, x2=0.0, v1=1.0, v2=0.0),
    Particle(2, x1=0.0, x2=4.0, v1=0.0, v2=1.0),
    Particle(3, x1=-2.12, x2=-2.12, v1=-0.7, v2=-0.7)
]

# Запуск алгоритма
best_pos, best_val = pso(particles, phi_p=1.5, phi_g=1.5, r_p=0.5, r_g=0.5)

print("\n\n=== ФИНАЛЬНЫЙ РЕЗУЛЬТАТ ===")
print(f"Лучшее решение: (x1={best_pos[0]:.4f}, x2={best_pos[1]:.4f}), f={best_val:.4f}")
```
